## Betting Odds Selector

```python
from spread_bettors_probs import spread_bettors_probs
import pandas as pd
import pickle
import warnings
import matplotlib.pyplot as plt

# load the starting dataframe
week7 = pd.read_excel('week_7.xlsx')

# load the regression model pickle file
filename = 'log_reg_spread_model.pkl'
loaded_model = pickle.load(open(filename, 'rb'))
```

```python
# run the model
spread_bettors_probs(week7, loaded_model, to_excel = True)
```

```python
week7 = week7.iloc[:2,:]
week7
```

```python
bills_bettors = [i/10 for i in range(3,10)]
sf_bettors = [1-i/10 for i in range(3,10)]
results= pd.DataFrame()
results['bills_bettors'] = bills_bettors
results['sf_bettors'] = sf_bettors
```

```python
for row in results.index:
    week7.loc[0, 'Bettors'] = results.loc[row, 'bills_bettors']
    week7.loc[1, 'Bettors'] = results.loc[row, 'sf_bettors']
    modeled_probs = spread_bettors_probs(week7, loaded_model)
    results.loc[row, 'prob_rating_bills'] = modeled_probs.loc[0, 'prob_rating']
    results.loc[row, 'prob_rating_sf'] = modeled_probs.loc[1, 'prob_rating']
results
```

```python
results.plot.line(x = 'bills_bettors', y = ['prob_rating_bills', 'prob_rating_sf'])
plt.ylabel('probability_rating')
```
