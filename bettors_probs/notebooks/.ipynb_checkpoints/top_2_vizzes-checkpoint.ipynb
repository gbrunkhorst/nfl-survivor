## Top 2 Picks Lookup Charts
Create lookups for selecting between the top 2 teams based on the fraction of bettors going with each. 

```python
import sys
path_to_working_script = r'..\bettors_probs'
sys.path.append(path_to_working_script)

import bettors_probs
import pandas as pd
import pickle
import warnings
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load the regression model pickle file.  I created this 
# previously in scikitlearn using the logisitic regression model
# converts spreads to probabilities
path = r'..\source_data'
file = 'log_reg_spread_model.pkl'
loaded_model = pickle.load(open(path+'\\'+file, 'rb'))
```

```python
def find_equilibrium(pa = None, pb = None, sa = None, sb = None, model = None):
    
    # create a BettorsProbs object
    bp = bettors_probs.BettorsProbs()
    
    #[ work generalizing to accept spreads]
    if pa is None:
        spreads = pd.Series([sa, sb])
        probs = bp.probs_spreads(spreads = spreads, model = model)
        probs = pd.Series(probs)
    else:
        probs = pd.Series([pa, pb])
    
    ba = 50
    bb = 50
    bettors = pd.Series([ba, bb])
    
    bet_strength = bp.bet_strength(bettors  = bettors, probs = probs)[0]
    
    while bet_strength > 1:
        bettors = pd.Series([ba, bb])
        bet_strength = bp.bet_strength(bettors  = bettors, probs = probs)[0]
        ba += 1
        bb += -1
    while bet_strength < 1:
        bettors = pd.Series([ba, bb])
        bet_strength = bp.bet_strength(bettors  = bettors, probs = probs)[0]
        ba += -.1
        bb += +.1

    bettors = pd.Series([ba, bb])
    bet_strength = bp.bet_strength(bettors  = bettors, probs = probs)[0]

    return ba, bet_strength
```

```python
# test the function
pa = .8
pb = .8

find_equilibrium(pa, pb)
```

## Make the chart based on odds

```python
# iterate to make the chart
z_df = pd.DataFrame()
pas = []
pbs = []
eqs = []
n  = 100
for pb in range(int(.5*n),n):
    pb = pb/n
    for pa in range(int(pb*n),n):
        pa = pa/n
        pas.append(pa)
        pbs.append(pb)
        eq = find_equilibrium(pa, pb)[0]
        z_df.loc[pb,pa] = eq
        eqs.append(eq)
```

```python
X, Y = np.meshgrid(z_df.columns.tolist(), z_df.index.tolist())
Z = z_df.values
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline=1, fontsize=10, fmt='%1.0f')
ax.set_title('Break-Even Point for Top Two Bets (Percent of Bets)')
ax.set_xlim(0.5, 1)
ax.set_ylim(0.5, 1)
ax.grid()
ax.set_xlabel('P Favorite')
ax.set_ylabel('P Second Favorite')
```

## Make the chart based on spreads

```python
# let's not look a ton of runtime warnings
%%capture --no-display
# iterate to make the chart
z_df = pd.DataFrame()
sas = []
sbs = []
eqs = []
n  = 1
start = -1
end = -30
stride = -1

for sb in range(start, end, stride):
    sb = sb/n
    for sa in range(int(sb), end, stride):
        sa = sa/n
        sas.append(sa)
        sbs.append(sb)
        eq = find_equilibrium(sa = sa, sb = sb, model = loaded_model)[0]
        z_df.loc[sb,sa] = eq
        
        eqs.append(eq)
```

```python
X, Y = np.meshgrid(z_df.columns.tolist(), z_df.index.tolist())
Z = z_df.values
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline=1, fontsize=10, fmt='%1.0f')
ax.set_title('Break-Even Point for Top Two Bets (Percent of Bets)')
ax.set_xlim(end, start)
ax.set_ylim(end, start)
ax.grid()
ax.set_xlabel('Spread Favorite')
ax.set_ylabel('Spread Second Favorite')
ax.invert_yaxis()
ax.invert_xaxis()
```
